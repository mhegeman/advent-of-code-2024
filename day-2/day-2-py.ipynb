```python
import pandas as pd
import numpy as np
max_columns =  10
column_names = [f'col{i}' for i in range(max_columns)]    
all_data = pd.read_csv('day-2-input.txt', names=column_names,delim_whitespace=True,header=None)
```

```python
def getminmax(df):
    df = df.agg(['max','min'],axis=1)
    return(df)

#takes the result of minmax
def test(df):    
    s = np.where(df['max']>0,
            np.where(df['min']>0, np.where(df['max']<4, 1, 0),0),
            np.where(df['min']>-4, np.where(df['max']<0, 1, 0),0)
        )
    return(s)
```

```python
#minmax=all_data.diff(axis=1).agg(['max','min'],axis=1)
getdiff = all_data.diff(axis = 1)
minmax = getminmax(getdiff)
t = test(minmax)
smoosh = {'orig': t}
t.sum()
```

```python
col_list = ['col1', 'col2', 'col3', 'col4', 'col5', 'col6', 'col7', 'col8', 'col9']
for column in col_list:
    #print(column)
    dropdf = all_data.drop(column, axis=1)
    diffdf = dropdf.diff(axis = 1)
    f = getminmax(diffdf)
    d = test(f)
    smoosh[column] = d

smoosh
```
