```python
def evaluate_expression(nums, ops):
    result = nums[0]
    for i, op in enumerate(ops):
        if op == '*':
            result *= nums[i + 1]
        elif op == '+':
            result += nums[i + 1]
        elif op == '||':
            result = int(str(result) + str(nums[i + 1]))
    return result

def find_valid_equations(filename):
    valid_equations = []
    
    with open(filename, 'r') as file:
        for line in file:
            target, numbers = line.strip().split(':')
            target = int(target)
            nums = [int(x) for x in numbers.strip().split()]
            
            n = len(nums) - 1
            found_solution = False
            
            # Generate all possible operator combinations
            ops_list = ['+', '*', '||']
            for ops in itertools.product(ops_list, repeat=n):
                if found_solution:
                    break
                
                result = evaluate_expression(nums, ops)
                if result == target:
                    equation = str(nums[0])
                    for i, op in enumerate(ops):
                        equation += f' {op} {nums[i+1]}'
                    equation += f' = {target}'
                    valid_equations.append(equation)
                    found_solution = True
    
    return valid_equations

import itertools

# Test with the sample input
valid_equations = find_valid_equations('day-7-input.txt')
total = 0
for equation in valid_equations:
    print(equation)
    total += int(equation.split('=')[1].strip())
print(f"\nSum of valid equations: {total}")
```
